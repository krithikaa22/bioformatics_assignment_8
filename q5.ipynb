```python
from Bio import SeqIO

seq = []
for seq_record in SeqIO.parse(open("Q4.fasta"), "fasta"):
    seq.append(str(seq_record.seq))
```

```python
import re

print("[SV]-T-[VT]-[DERK]-{IL}")
for i in seq:
    print(">",i)
    for match in re.finditer("[SV]T[VT][DERK]{2}[^IL]", i):
        print(match.group(),' - ', match.span())
    print("\n")
print("\n")
print("[FILV]Qxxx{RK}Gxxx[RK]xx[FILVWY]")
for i in seq:
    print(">",i)
    for match in re.finditer("[FILV]Q...[^RK]G...[RK]..[FILVWY]", i):
        print(match.group(),' - ', match.span())
    print("\n")
```
