```python
def window_plot(hydro_seq,w):
    y = []
    for i in range(w):
        y.append(hydro_seq[i])
    for i in range(w,len(hydro_seq)-w):
        y.append((sum(hydro_seq[i-w:i])+sum(hydro_seq[i:i+w+1]))/(2*w))
    for i in range(len(hydro_seq)-w,len(hydro_seq)):
        y.append(hydro_seq[i])

    plt.plot([i for i in range(len(hydro_seq))], y)

window_plot([hydrophobicity[i] for i in seq], 4)
```
